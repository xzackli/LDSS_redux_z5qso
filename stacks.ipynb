```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
```

```python
filenames = ['63', '64', '65', '70', '71', '72', '77', '78', '79']
filenames = ['ccd00' + f + '_2spec.fits' for f in filenames]
```

```python
stack = []
masks = []
for filename in filenames:
    plt.figure()
    plt.gca().axhline(0.0,color='r')
    image_data, mask = fits.getdata(filename, ext=0)
    plt.plot( np.mean(image_data,axis=1))
    plt.title(filename.replace('_', '\_'))
    stack.append(image_data)
    masks.append(mask)
```

```python
stacked = np.sum(np.array(stack),axis=0)
```

```python
# plt.figure(dpi=200)
plt.imshow( np.arcsinh(stacked[:,500:700]), origin='lower' )
```

```python
plt.plot( np.sum(stacked[:,500:700],axis=1) )
plt.gca().axhline(0,color='red')
```

```python
plt.figure(dpi=200)
plt.imshow( np.arcsinh(stack[1]*masks[1]), origin='lower' )
```

```python
plt.figure(dpi=200)
plt.imshow(np.arcsinh(stack[1]), origin='lower')
plt.figure(dpi=200)
plt.imshow(masks[1], origin='lower')
```
